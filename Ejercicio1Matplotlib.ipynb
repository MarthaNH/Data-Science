```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(1, 11, 0.1)
z = np.random.randint(5,10,100) 

plt.figure(figsize=(8,8))
ax1 = plt.subplot(3,3,(1,3))
y1 = x**3 + 2* np.random.randn(100)
ax1.scatter(x, y1, c = z, cmap = 'plasma')
ax1.plot(x, x**3 + 2, '--', color = 'orange', label = '$x^3 + 2$')
ax1.legend()
ax1.grid(True)

ax2 = plt.subplot(3,3,(4,5))
ax2.plot(x, x**1, '*', color = 'pink', label = '$x$')
ax2.plot(x, x**2, 'o', color = 'blue', label = '$x^2$')
ax2.legend()

plt.xlim(0,10) # limites del eje x
plt.ylim(0,100) # limites del eje y
ax2.grid(True)

ax3 = plt.subplot(3,3,7)
y3 = np.random.randn(100)*0.5
ax3.plot(x, y3, '^', ms = 5, lw=3 , mec ='yellow', mew=2, label = '$4x^2$')



ax4 = plt.subplot(3,3,8)
y4 = [np.random.normal(0, std, 20) for std in range(1, 4)]
plt.boxplot(y4,vert=True);  

ax5 = plt.subplot(3,3,(6,9))
ax5.plot(x, x**3, '*', color = 'red', ms = 5, lw=3, label = '$3x^3$', alpha=0.2)
ax5.grid(True)

axes = [ax1, ax2, ax3, ax4, ax5]
```
