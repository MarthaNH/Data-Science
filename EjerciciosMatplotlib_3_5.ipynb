Ejercicio 3. Colocar en cero los valores en rojo de la imagen de la luna, hacer lo mismo para los valores en verde y luego el azul 


```python

import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

```python
imagen = Image.open('/content/Full_Moon_Luc_Viatour.jpg') 
im_datos = np.array(imagen)
x, y, z = im_datos.shape

im_r = np.array(imagen)
im_v= np.array(imagen)
im_a= np.array(imagen)

for i in range (x):
    for j in range (y):
        im_r[i][j][0] = 0
        im_v[i][j][1] = 0
        im_a[i][j][2] = 0
                        

fig,(ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(14,14))
ax1.imshow(im_r)
ax1.axis(False)
ax2.imshow(im_v)
ax2.axis(False)
ax3.imshow(im_a)
ax3.axis(False)
```

Ejercicio 4. Transformar una imagen a escala de grises


```python

imagen = Image.open('/content/cat.jpg') 
im_datos = np.array(imagen)
x, y, z = im_datos.shape

im_gris = np.array(imagen)

for i in range (x):
    for j in range (y):
        maximum = im_gris[i][j].argmax()
        for k in range(z):
            if not k == maximum:
                im_gris[i][j][k] = im_gris[i][j][maximum]
  
        
fig, (ax1, ax2)= plt.subplots(1, 2, figsize=(12,12))
ax1.imshow(im_datos)
ax1.axis(False)
ax2.imshow(im_gris)
ax2.axis(False)
```
